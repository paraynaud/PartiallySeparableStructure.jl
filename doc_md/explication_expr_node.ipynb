 - Nécessité de lire le fichier concernant l'architecture globale avant de lire ce fichier.

# Objectifs : 
 - Ce fichier à pour but d'expliquer le contenu du répertoire node_expr_tree
 - Abstraction des noeuds opérations peut-importe la structure d'arbre d'expression derrière.
 - Mes algorithmes travaillent sur les noeuds définis par ces structures que ce soit Expr ou bien les arbres que j'ai fais.
 
 

# Explications : 

D'après l'expérience de mon stage je savais que je devrais traiter des arbres d'expressions, par conséquent je souhaitais être capable de manipuler différentes structures ayant le comportements d'arbres d'expressions. Sachant que je partais de JuMP, je souhaitais tout d'abord traiter la structure Expr (récupérée à l'aide de MOI), mais d'un point de vu algorithmique il suffisait de traiter arbres avec des noeuds représentant une opération, une variable ou une constante. 

Il était donc nécessaire de créer les noeuds représentant les opérations suivant ma structure d'arbres établie précédemment. Je savais plus ou moins ce qu'il me fallait mais encore une fois pour rester souple j'ai tout d'abord défini un type abstrait qui engloberait les types réels que j'utiliserai ensuite.
J'avais donc besoin de définir : 
 - une type abstrait
 - un type pour les constantes 
 - un type pour les variables 
 - un type pour les opérations ne prenant pas de paramètre (+, -, x)
 - un type pour les opérations plus compliqué prenant des paramètres ( ^n)
 
Ce qui correspond respectivement aux fichiers : 
 - abs_expr_node.jl 
 - impl_constants.jl
 - impl_variables.jl
 - impl_simple_operator.jl
 - impl_complex_omperator.jl
 
Comme pour chaque répertoire il y a également une interface récapitulant les fonctions que les modules satisfaisant le trait seront amené à implémenter : 
 - itf_expr_node.jl 
 
Et également la définition du trait : 
 - tr_expr_node.jl



# Fonction requise par le constructeur et l'interface : 

Le constructeur requiert une fonction capable de créer chaque type de noeud différent que j'ai implémenté, appelé selon un fonction ayant le même nom mais diffénrencié par les arguments: 
 - create_node_expr

Ensuite pour garder un niveau d'abstraction quand j'aurais besoin de manipuler les arbres contenant ces noeuds il était nécessaire d'avoir des fonctions simples : 
 - \_node_is_operator
 - \_node_is_variable
 - \_node_is_constant
 
Il fallait ensuite préciser les opérations il fallait difféncier les différentes opérations: 
 - \_node_is_plus
 - \_node_is_minus
 - \_node_is_times
 - \_node_is_power
 - \_node_is_power
 - \_node_is_sin
 - \_node_is_cos
 - \_node_is_tan

On peut considérer les fonctions précédentes comme étant des fonctions de base pour l'utilisation des noeuds, cependant par la suite nous avons besoin de typer un arbre d'expression, et les règles dépendent des noeuds d'opérations qu'elles traversent. L'algorithme réalisant le typage d'un arbre n'est donc pas défini dans ce répertoire, étant donné que nous ne parlons pas d'arbre dans ce répertoire, mais les règles de typage qui dépendent des noeuds sont elles implémenté dans ce répertoire : 
 - \_get_type_node

Cette fonction type le noeud en fonction du type de noeud (opération/variable/constante) et le type des ces enfants qui on déjà été traité auparavant.
Il est intéressant de noter que historiquement ce répertoire est présent depuis le début du projet. J'ai ajouté par la suite le typage d'arbre d'expressions, et par conséquent les type d'expression (linéaire/constants/plus). 
Le typage à créé une dépendance entre ces deux répertoires, donc **node_expr_tree** dépend de **type_expr**. 

Je définirai plus tard un graph de dépendance entre les répertoires.
